## Dlaczego warto się męczyć

JUPYTER to połączenie MsWord/LibreOfice z RMarkdown/ShinyR oraz R (albo jeśli wolicie Python-a, JAVA, itp ). Na dodatek wszystko to działa w przeglądarce, więc trochę przypomina Google Docs. Zabawa z edytorem składa się z następujących etapów:

   1. uruchomiamy JUPYTERA - w Windowsie można to zrobić z wiersza poleceń: _jupyter notebook_
   2. w przeglądarce otwiera się nowa strona, na której wyświetla się robocza lokalizacja (czyli folder) - tutaj domyślnie zapisywane są nasze pliki wyprodukowane przy użyciu JUPYTERA.
   3. Wybieramy sobie jakąś lokalizację i klikamy tworzenie nowego pliku JUPYTER. 
   4. Otwiera się nowa strona w przeglądarce z włączonym edytorem. To bardzo przypomina Google Docs - biała strona, a u góry różne opcje edycji.
   4. Na naszej białej stronie możemy dodawać różne rodzaje akapitów (nagłówki, zwykły tekst, kod R/Python)
   5. Żeby zatwierdzić kod/tekst naciskamy przycisk "Run" (jest też jakiś skrót klawiszowym, ale nie pamiętam jaki)
   6. Zatwierdzony kod/tekst lub kod R/Python jest wykonywany - tekst zmienia się w tekst, obliczenia w wynik lub grafikę
   7. Koniec.

W efekcie powstaje coś na kształt prezentacji, którą możecie w czasie rzeczywistym zmieniać. Jeśli nie pasuje Wam jakiś akapit możecie go w każdej chwili kliknąć i poprawić. ## Schody (dla użytkownika Windowsa)

   1. Instalacją pakietów w Pythonie zajmuje się pakiet PyPi (od _Python Package Index_ - strona na której są wrzucane wszystkie pakiety, coś w rodzaju R-owego CRANa) zwany również pip, który niestety też musiałem zainstalować.
   2. Instalacja pip polega na ściągnięciu pliku _[get-pip.py](https://pip.pypa.io/en/latest/installing/)_ i uruchomieniu go przy pomocy Pythona. Dodatkowo trzeba jeszcze "podpiąć" pip pod wiersz poleceń Widowsa, co nie jest trywialne (wszystko zostało opisane [tutaj](https://www.youtube.com/watch?v=zPMr0lEMqpo)   )
   3. korzystanie z pip polega na podawaniu mu poleceń przy pomocy wiersza poleceń Windowsa - mało wygodne (w porównaniu z R)
   4. niestety w moim przypadku (64 bitowy Windows 7 Python 2.7) nie wystarczy wpisać __pip install numpy__ w wierszu poleceń - konieczne jest zastosowanie następującej procedury:
       * Instalacja dodatku do Windowsa pt. [Microsoft Visual C++ Compiler for Python 2.7 ](https://www.microsoft.com/en-us/download/details.aspx?id=44266)
       * wejście na stronę Christopha Gohlke, gdzie znajdują się nieoficjalne wersje różnych pakietów do samodzielnej instalacji (to niepoważne, że istotny element oprogramowania został przygotowany nieoficjalnie i nie jest dostępny na głównej stronie Pythona albo PyPi)
       * pobranie [ze strony Christopha Gohlke](http://www.lfd.uci.edu/~gohlke/pythonlibs/) odpowiedniego pliku z rozszerzeniem .whl (skrót od "wheel" - Python od jakiegoś czasu stosuje pliki .whl zamiast .exe) - w moim przypadku był to pakiet _numpy_
       * uruchomienie nieoficjalnego pliku z pakietem przy pomocy wiersza poleceń i komendy __pip install ścieżka_dostępu\numpy__
   5. Wreszcie można korzystać z załadowanych pakietów.

To jest skrócony opis problemów. Warto dodać, że dotyczą one tylko niektórych bibliotek (tych kluczowych). Oczywiście większość ww. komplikacji można obejść instalując przy użyciu dystrybucji [Anaconda](https://www.continuum.io/why-anaconda) Pythona 3.4 wyposażonego we wszystkie niezbędne biblioteki/narzędzia  albo zamieniając Windowsa na Linuxa. W moim wypadku żadna z tych opcji nie wchodziła w grę (chciałem pójść po największej linii oporu) i dlatego musiałem się męczyć.

```python
%matplotlib inline # to polecenie mowi JUPYTERowi ze grafike wyprodukowana przez pakiet matplotlib ma pokazac "w lini kodu" - bez tego polecenia grafika otwiera sie w oddzielnym oknie

from mpl_toolkits.mplot3d import Axes3D
from matplotlib.collections import PolyCollection
from matplotlib.colors import colorConverter
import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')


def cc(arg):
    return colorConverter.to_rgba(arg, alpha=0.6)

xs = np.arange(0, 10, 0.4)
verts = []
zs = [0.0, 1.0, 2.0, 3.0]
for z in zs:
    ys = np.random.rand(len(xs))
    ys[0], ys[-1] = 0, 0
    verts.append(list(zip(xs, ys)))

poly = PolyCollection(verts, facecolors=[cc('r'), cc('g'), cc('b'),
                                         cc('y')])
poly.set_alpha(0.7)
ax.add_collection3d(poly, zs=zs, zdir='y')

ax.set_xlabel('X')
ax.set_xlim3d(0, 10)
ax.set_ylabel('Y')
ax.set_ylim3d(-1, 4)
ax.set_zlabel('Z')
ax.set_zlim3d(0, 1)

plt.show()
```
